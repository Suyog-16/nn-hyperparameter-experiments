```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_learning_rate = pd.read_csv("results/learning_rates.csv")
df_learning_rate
```

```python
plt.figure(figsize=(8,5))
plt.plot(df_learning_rate['Learning Rate'],df_learning_rate['Average Loss'],color = 'red')
plt.title("Learning Rate vs Average Loss")
plt.xlabel("Learning Rate")
plt.ylabel("Average loss")
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df_learning_rate['Learning Rate'],df_learning_rate['Accuracy'],color = 'red')
plt.title("Learning Rate vs Accuracy")

plt.xlabel("Learning Rate")
plt.ylabel("Accuracy")
plt.xticks([0.01,0.1,0.2,0.3,0.4,0.9])
plt.show()
```

```python
df_batch = pd.read_csv("results/batch_size.csv")
df_batch
```

```python
plt.figure(figsize=(8,5))
plt.plot(df_batch['Batch'],df_batch['Accuracy'],color = 'red')
plt.title("Batch Sizes vs Accuracy")
plt.xlabel("Batch Sizes")
plt.ylabel("Accuracy")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(df_batch['Batch'],df_batch['Average Loss'],color = 'red')
plt.title("Batch Sizes vs Average Loss")
plt.xlabel("Batch Sizes")
plt.ylabel("Avg Loss")
plt.show()
```
